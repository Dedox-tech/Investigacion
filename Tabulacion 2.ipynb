# 1. Extracción de datos

```python
#Librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Lectura de los datos
dataset = pd.read_csv(filepath_or_buffer = "C:/Diego/Datos mendeley 7.csv", sep = ",")
```

```python
#Vista
dataset.head(n=5)
```

```python
#Columnas
print(dataset.columns)
```

```python
#Informacion adicional
dataset.info()
```

```python
#Elegir las variables claves
dataset_processed = dataset[["Key", "Item Type", "Publication Year", "Author", "Title", "Publication Title", "DOI", "Abstract Note", "Date", "Pages", "Issue", "Volume", "Manual Tags", "Extra"]]
```

```python
dataset_processed.head(n=196)
```

# 2. Años de publicación

```python
#Graficar los años de publicación
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,5))
sns.set(context = "notebook", style = "ticks")
dataset_processed["Publication Year"].value_counts().sort_values(ascending=False).nlargest(7).plot(kind = "bar")
fig.tight_layout()
```

# 3. Revistas donde se publican los articulos

```python
#Graficar las revistas
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,7))
sns.set(context = "notebook", style = "ticks")
dataset_processed["Publication Title"].value_counts().nlargest(7).plot(kind = "bar")
fig.tight_layout()
```

# 4. Editores que publican los articulos

```python
#Graficar los editores
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,7))
sns.set(context = "notebook", style = "ticks")
dataset_processed["Extra"].value_counts().nlargest(5).plot(kind = "bar")
fig.tight_layout()
```

# 5. Palabras más utilizadas en los titulos

```python
#Graficar las plabras mas utilziadas en los titulos
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,5))
sns.set(context = "notebook", style = "ticks")
dataset_processed["Title"].str.split(expand=True).stack().value_counts().nlargest(10).plot(kind = "bar")
fig.tight_layout()
```

# 6. Palabras claves (Keywords) más utilizadas 

```python
#Graficar las keywords mas utilizdas
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (7,5))
sns.set(context = "notebook", style = "ticks")
dataset_processed["Manual Tags"].str.split(expand=True).stack().value_counts().nlargest(10).plot(kind = "bar")
fig.tight_layout()
```
